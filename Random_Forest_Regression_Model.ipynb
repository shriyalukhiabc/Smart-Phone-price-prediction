```python
import pandas as pd
data=pd.read_csv('/content/sample_data/final_smartphones.csv')
df=pd.DataFrame(data)
display(df)
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
x=df.drop(columns='price')
y=df['price']
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.20,random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
```

```python
step1=ColumnTransformer(remainder='passthrough',transformers=[('cols',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False),[0,2,3,4,5,6,11,19])])
```

```python
from sklearn.ensemble import RandomForestRegressor
step2=RandomForestRegressor(n_estimators=10, random_state=0, oob_score=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(step1,step2)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy = pipe.score(x_test,y_test)
print('Accuracy of the Model is:- ',accuracy*100)
```

```python
from sklearn.metrics import mean_absolute_error
mse = mean_absolute_error(y_pred,y_test)
print('Mean Absolute Error is :- ',mse)
```
